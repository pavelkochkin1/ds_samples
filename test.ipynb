```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
```

```python
data = pd.read_csv('iris.data', names=('sepal_lenght','sepal_width','petal_length','petal_width','kind'))
data.head()
```

```python
data.kind.unique()
```

```python
data.kind = data.kind.map({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})
```

```python
from pandas.plotting import scatter_matrix
colors = ['red', 'green', 'blue']
scatter_matrix(data, diagonal='kde', figsize=[10,10], c=data.kind.apply(lambda x:colors[x]))
None
```
